```python
import pandas as pd
import json

rest_origin = pd.read_json("../../No3Project/shop1",encoding="utf-8")

rest = rest_origin[ ["bizhour", "href", "name", "phone", "rating"] ]
```

```python
def care(x):
    if len(x) < 2:
        if x == "冰":
            return False
        return True
rest["Care"] = rest["name"].apply(care)
```

```python
rest = rest[rest["Care"] != True]
```

```python
rest = rest[rest["name"] != "123"]
```

```python
rest = rest[(rest["phone"] != "Not Available") & (rest["rating"] != 0) & (rest["bizhour"] != "暫無提供")] 
```

```python
rest_list = []
for name in rest["name"]:
    rest_list.append(name)
```

```python
ptt_origin = pd.read_json("../../No3Project/ptt_no_clean.json",encoding="utf-8")
ptt_origin
```

```python
def extract_name(x):
    for name in rest_list:
        if x.find(name) != -1:
            return name
ptt_origin["res_name"] = ptt_origin["content"].apply(extract_name)
```

```python
def tp(x):
    if x.find("台北") != -1:
        return True
    elif x.find("新北") != -1:
        return True
    elif x.find("臺北") != -1:
        return True

ptt_origin["TPorN"] = ptt_origin["content"].apply(tp)
```

```python
ptt_origin["TPorNinT"] = ptt_origin["title"].apply(tp)
```

```python
ptt_origin
```

```python
ptt_origin = ptt_origin[(ptt_origin["TPorN"] == True) | (ptt_origin["TPorNinT"] == True)]
```
