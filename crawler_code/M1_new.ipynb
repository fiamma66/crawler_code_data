```python

def extract_phone(x):
    phone = x[0]["phone"]
    phone = phone.replace("(","").replace(")","").replace("-","").replace("－","").replace("）","").replace("（","")
    phone = phone.translate(str.maketrans('０１２３４５６７８９','0123456789'))
    return phone

m1_ori["phone"] = m1_ori["info"].apply(extract_phone)
```
