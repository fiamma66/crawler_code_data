```python
import pandas as pd
import json
import jieba
import re
```

```python
m1 = pd.read_json("e:/pythonProject/crawler_code/cb104_G3.Mobile01.json",encoding="utf-8")
```

```python
m1 = m1.drop(["_id","href"],axis=1)
```

```python
def extract_info(lis):
    return lis[0]["place"]
        

m1["place"] = m1["info"].apply(extract_info)
```

```python
groupPlace = m1.groupby("place").apply(lambda x: [",".join(x["content"])])
```

```python
m1
```

```python
df = pd.DataFrame({"place":groupPlace.index, "content":groupPlace.values})
```

```python
df["content_list"] = df["content"].apply(lambda x: x[0].split(","))
```

```python
df["content_list"][0]
```

```python

from textrank4zh import TextRank4Keyword, TextRank4Sentence

text = df["content_list"][44][0]
tr4w = TextRank4Keyword()
tr4w.analyze(text=text, lower=True, window=2)
for item in tr4w.get_keywords(10, word_min_len=2):
    print(item.word, item.weight)

tr4s = TextRank4Sentence()
text = " ".join(jieba.cut(text))
print("====================")
print("關鍵句:")
tr4s.analyze(text=text, lower=True, source="no_stop_words")
# 無法使用 textrank4zh sentence 使用標點符號斷句
# 目前資料已清除非文字
for sentence in tr4s.get_key_sentences(num=3):    
    print(sentence.weight, sentence.sentence)
```
